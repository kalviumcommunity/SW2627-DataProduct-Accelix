# GitHub Repository & Team Workflow Setup

This lesson explains how a data team should work together in a shared repository: branching, issue tracking, pull requests, code review, and commit message conventions.

## Why Branch Strategy Matters

Branching keeps work isolated so one teammate's changes do not break another teammate's work. `main` should stay releasable, while feature branches hold in-progress work until it is ready to merge.

A branch is a separate line of development. Teams use branches to work in parallel, review changes safely, and make rollbacks easier if something goes wrong.

Branch names should communicate intent. A name like `feature/churn-model` is much clearer than a vague name like `my-new-code`.

## GitHub Issues and Sprint Task Tracking

A GitHub issue is a trackable unit of work. It should include a clear title, enough context to understand the task, labels to describe the work type, and an assignee who is accountable for completion.

Good issues make work visible. They create a record of why a change was made and help the team understand what needs to happen next.

## Pull Requests and Code Review

A pull request is the control point between a feature branch and `main`. It gives teammates a chance to review the code before it is merged.

A typical workflow looks like this:

```text
1. Create a feature branch
2. Make changes locally
3. Push the branch to GitHub
4. Open a pull request
5. Link the related issue
6. Review and revise the code
7. Approve and merge
8. Delete the branch
```

Use phrases like `Closes #123` in the PR description so GitHub can connect the pull request to the issue automatically.

## Commit Message Conventions

Conventional commits make history easier to read and automate. The format separates the type of change from the description.

```text
feat: add data validation schema checker
fix: correct null percentage calculation in profiler
docs: update README with dataset intake process
refactor: extract column renaming logic into utils
test: add unit tests for date parser
```

Common types include `feat`, `fix`, `docs`, `refactor`, and `test`. The goal is to explain intent, not just activity.

## Team Workflow Summary

A healthy repository workflow uses branches for isolation, issues for planning, pull requests for review, and commit messages for history. Together, these practices make work safer, clearer, and easier to maintain across a team.
