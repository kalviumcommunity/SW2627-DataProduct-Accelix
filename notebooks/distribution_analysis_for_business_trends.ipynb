# Distribution Analysis for Business Trends

Understanding the shape of a distribution is often more important than computing a single average. Skewness, kurtosis, multimodality, and segment differences can all reveal business patterns that summary statistics hide.

## The Real Scenario

A revenue dataset may report a mean of 5,000, but that number can be misleading if the distribution is heavily skewed. If most customers spend under 50,000 and a small number of enterprise accounts spend far more, the mean gets pulled upward by outliers. In that case, the median is usually a more representative measure of a typical customer.

## Understanding Distributions

### Skewness

Skewness measures whether a distribution is symmetric. A value near 0 is approximately symmetric. Positive skew means the right tail is longer, which often happens when a few customers or orders are much larger than the rest. Negative skew means the left tail is longer. When a distribution is strongly skewed, the median is usually more reliable than the mean.

### Kurtosis

Kurtosis describes how heavy the tails are and how often extreme values appear. High kurtosis suggests more outliers and more risk that unusually large or small values are affecting the business metric.

### Computing Skewness and Kurtosis

```python
from scipy import stats

skewness = stats.skew(df['revenue'])
kurtosis = stats.kurtosis(df['revenue'])

print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")

if abs(skewness) > 1:
    print("Highly skewed - use median")
if kurtosis > 3:
    print("Heavy tails - expect outliers")
```

This check helps an analyst decide whether the mean is representative or whether the median and robust segmentation are better business choices.

## Visualizing and Comparing Distributions

### Histogram

A histogram groups values into buckets and shows how many observations fall into each bucket. It is the fastest way to see whether a metric is centered, skewed, or possibly multimodal.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(df['revenue'], bins=50, edgecolor='black')
plt.xlabel('Revenue')
plt.ylabel('Count')
plt.title('Revenue Distribution')
plt.show()
```

### KDE

A kernel density estimate smooths the distribution so the overall shape is easier to read without bucket noise.

```python
df['revenue'].plot(kind='density', figsize=(10, 5))
plt.xlabel('Revenue')
plt.title('Revenue Distribution (Smoothed)')
plt.show()
```

### Comparing Segments

Segment comparison is where distribution analysis becomes directly useful for business decisions. If high-value customers and low-value customers have visibly different revenue distributions, they may need different pricing, retention, or sales strategies.

```python
high_value = df[df['revenue'] > df['revenue'].quantile(0.75)]
low_value = df[df['revenue'] < df['revenue'].quantile(0.25)]

plt.figure(figsize=(10, 5))
plt.hist(high_value['revenue'], alpha=0.5, label='High-Value', bins=30)
plt.hist(low_value['revenue'], alpha=0.5, label='Low-Value', bins=30)
plt.legend()
plt.show()
```

If the combined distribution looks bimodal, that can indicate two business populations rather than one. Reporting one average for both groups may hide that difference and lead to poor forecasts.

## Key Takeaways

Always inspect distributions before relying on averages. Skewness tells you when the median is safer than the mean. Kurtosis warns about extreme outliers. Histograms and KDE plots reveal shape, while segment comparisons show whether hidden business groups exist. A good analyst reports what the data actually looks like, not just what a single summary number suggests.
