# Development Environment & Workspace Setup

Before a single line of analysis is written, the workspace has to be reproducible. This lesson covers the minimum structure a data project needs so a new teammate can clone the repo, set up the environment, and run the work without guesswork.

## The Real Scenario

A team breaks a project because dependencies drift, paths differ across operating systems, and setup instructions live in Slack instead of the repo. The fix is a deliberately managed workspace: an isolated virtual environment, pinned dependencies, a predictable folder structure, a proper `.gitignore`, and a README that explains how to get started in a few commands.

## Why Isolated Environments Matter

A virtual environment keeps each project's packages separate from the global Python installation. That prevents one project from breaking another and makes it possible for teammates and CI to reproduce the same analysis environment.

A `venv` is just a self-contained directory created by Python's built-in `venv` module. When it is activated, `python` and `pip` point to that isolated location instead of the system installation.

## Creating and Activating a Python venv

Run these commands from the project root.

```bash
# macOS and Linux
python3 -m venv venv

# Windows
python -m venv venv
```

```bash
# macOS and Linux
source venv/bin/activate

# Windows Command Prompt
venv\Scripts\activate

# Windows PowerShell
venv\Scripts\Activate.ps1
```

```bash
deactivate
```

Always activate the environment before installing packages, or the installation will land in the global Python environment instead of the project.

## Designing a Data Project Folder Structure

A standard layout keeps raw data, processed data, notebooks, scripts, and outputs easy to find.

```text
customer-analytics/
├── data/
│   ├── raw/
│   └── processed/
├── notebooks/
├── scripts/
├── output/
├── venv/
├── .env
├── .gitignore
├── requirements.txt
└── README.md
```

`data/raw/` stores source files exactly as received. `data/processed/` stores cleaned data ready for analysis. `notebooks/` holds exploratory and reporting notebooks. `scripts/` contains repeatable automation. `output/` stores generated figures, exports, and reports.

## Managing Dependencies with requirements.txt

`requirements.txt` is the contract that lets another person recreate your environment. It lists the packages and versions the project depends on.

```text
pandas==2.1.4
numpy==1.26.2
matplotlib==3.8.2
seaborn==0.13.0
jupyter==1.0.0
scikit-learn==1.3.2
python-dotenv==1.0.0
openpyxl==3.1.2
```

Capture the current environment with `pip freeze > requirements.txt` after installing or updating packages. A teammate can then clone the repo, create a venv, activate it, and run `pip install -r requirements.txt`.

## Version Control Hygiene with .gitignore

`.gitignore` keeps machine-specific, secret, and generated files out of version control. Typical entries include the virtual environment, `.env`, Python caches, notebook checkpoints, and regenerable outputs.

```text
venv/
.venv/
.env
*.key
*.pem
__pycache__/
*.pyc
.ipynb_checkpoints/
.DS_Store
output/*.csv
output/*.pdf
dist/
build/
```

The critical rule is simple: never commit secrets or the `venv/` folder. Both are unsafe and unnecessary in source control.

## README Documentation for Team Handoff

A README should answer four questions immediately: what the project does, how to set up the environment, what each folder is for, and how to run the analysis. If a new teammate has to ask those questions, the README is incomplete.

A practical README setup section looks like this:

```text
git clone https://github.com/team/customer-analytics.git
cd customer-analytics
python3 -m venv venv
source venv/bin/activate
pip install -r requirements.txt
```

Include a `.env.example` file that lists required environment variables with placeholder values, so teammates know exactly what to configure without exposing secrets.

## Key Takeaways

A data project is easier to trust when the environment is isolated, dependencies are pinned, the folder structure is predictable, and the setup instructions are complete. If a new teammate can reproduce the project quickly from a clean clone, the workspace is doing its job.
